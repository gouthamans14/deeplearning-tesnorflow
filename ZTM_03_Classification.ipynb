```python
import numpy as np


def plot_decision_boundary(model,X,y):

  x_min,x_max = X[:, 0].min()-0.1, X[:, 0].max()+0.1
  y_min,y_max = X[:, 1].min()-0.1, X[:, 0].max()+0.1

  xx, yy = np.meshgrid(np.linspace(x_min, x_max,100),np.linspace(y_min, y_max,100))
  x_in = np.c_[xx.ravel(),yy.ravel()]

  y_pred = model.predict(x_in)

  if len(y_pred[0])>1:

    y_pred =np.argmax(y_pred,axis=1).reshape(xx.shape)
  else:
    y_pred = np.round(y_pred).reshape(xx.shape)

    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:, 0],X[:, 1],c=y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

```python
plot_decision_boundary(model=model_3,X=X,y=y)
```

# **   How to increase the learning rate / Finding the ideal learning rate**

Steps:
* A learning rate **Callback**
* Another model to use the callback
* A modified loss curve plots

## Model with Callback

```python
import tensorflow as tf
tf.__version__
```

```python
tf.random.set_seed(42)

model_9 = tf.keras.Sequential([
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

model_9.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer="Adam",
                metrics=['accuracy'])

# initializing the lr scheduler
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20) )

# fit the model with lr_scheduler

history_9 = model_9.fit(X_train,y_train, epochs=100, callbacks=[lr_scheduler])
```

# To build Multiclass Classification Model we need tweak few params

* 1) Input Shape = (28,28) -> Shape of one Image <break>
* 2) Output layer shape is one per class = 10 <break>
* 3) Output Layer Activation = Softmax (Not Sigmoid)
* 4) Loss function = Categorical Ctoss enntrophy
* 5) If your labels are integers -> Use sparsecategoricalcorssentropy
* 6) If your labels are one hot encoded(we can convert integrest to one hot encoded) -> Use Categorical Cross entropy
* 7) For binary classification its CategoricalCrossEntropy

```python
import random

def plot_random_image(model,images,true_labels,classes):

  """
  picks a random image, plots it and labels it with a prediction and truth label

  """
  #set up a random integer
  i = random.randint(0, len(images))

  target_image = images[i]
  pred_probs = model.predict(target_image.reshape(1,28,28))
  pred_label = classes[pred_probs.argmax()]
  true_labels = classes[true_labels[i]]

  plt.imshow(target_image,cmap=plt.cm.binary)

  if pred_label==true_labels:
    color= "green"
  else:
    color ="red"

  plt.xlabel("Pred: {} {:2.0f}%  True: {}".format(pred_label,100*tf.reduce_max(pred_probs),true_labels),
             color=color)

```

```python
plot_random_image(model=model_13,
                  images=test_data_norm,
                    true_labels=test_labels,
                  classes = class_names)
```
